```python
import pickle

# Function to load the inverted index from pickle file
def load_index_from_pickle(filename):
    with open(filename, 'rb') as f:
        inverted_index = pickle.load(f)
    return inverted_index

# Load the inverted index from pickle file
inverted_index = load_index_from_pickle("../scraper/inverted_index.pickle")

count = 0
for term, postings in inverted_index.items():
    print(f"Term: {term}")
    print("Postings:")
    for posting in postings:
        print(posting)
    print()
    count += 1
    if count == 10:
        break
```

```python
import pickle

# Load TF-IDF matrix and vectorizer from the pickle files
with open("tfidf_matrix.pickle", "rb") as f:
    tfidf_matrix = pickle.load(f)

with open("tfidf_vectorizer.pickle", "rb") as f:
    vectorizer = pickle.load(f)

# Print the vocabulary
print("Vocabulary:")
print(vectorizer.vocabulary_)

# Iterate over the non-zero elements in the TF-IDF matrix and print them with their corresponding terms
for row, col in zip(*tfidf_matrix.nonzero()):
    term = list(vectorizer.vocabulary_.keys())[list(vectorizer.vocabulary_.values()).index(col)]
    tfidf_score = tfidf_matrix[row, col]
    print(f"Document {row}, Term '{term}': TF-IDF score {tfidf_score}")
```
